Plotting a graph using the equation of *\2sin(theta)cos(theta)* using pytorch arange()

```python
x_c = torch.arange(1,1000,10)
y_c = 2*np.sin(x_c)*np.cos(x_c)
plt.plot(x_c,y_c)
```

```python
p_c = np.linspace(1,10,1000)
o_c = 2*np.sin(p_c)*np.cos(p_c)
plt.plot(x_c,y_c,p_c,o_c)
```

**Mutable and immutable array**

```python
index = 0
size = 10
value = 5
x = np.arange(size)
print(x)
x[index] = value
x
```

```python
# **jax depends on functional programming, so arrays are immutable, it only replaces the previous array**

index = 0
size = 10
value = 5
x = jnp.arange(size)
print(x)
x = x.at[index].set(value)
print(x)
```
